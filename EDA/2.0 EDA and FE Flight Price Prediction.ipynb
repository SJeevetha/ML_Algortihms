Feature Engineering

```python
## we just need to split the date,month,year from "Date_of_Journey" column
df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
## To convert the object[i.e string] into numerical[i.e int]
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop date of journey
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
## To remove the feature like "01:10 22 Mar" to "01:10" in Arrival_Time column
df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
## Split the arrival time into hours and minutes
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minutes'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
## To convert the object[i.e string] into numerical[i.e int]
df['Arrival_hour'] = df['Date'].astype(int)
df['Arrival_minutes'] = df['Month'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minutes'] = df['Dep_minutes'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
## since there is only one record that has null value
```

```python
## replace that null value with "1 stop" and also replace others with ordinal labels
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df[df['Total_Stops'].isnull()]
## no more null values there
```

```python
df.head(2)
```

```python
## We can delete the column "Route" because we already have seperate columns of "source and destination"
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
## df['Dur_hr'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
## df['Dur_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Dur_hr'] = df['Duration'].apply(lambda x: x.split('h')[0] if 'h' in x else '0')
df['Dur_min'] = df['Duration'].apply(lambda x: x.split('m')[0].split('h')[-1] if 'm' in x else '0')
```

```python
df['Dur_hr'] = df['Dur_hr'].astype(int)
df['Dur_min'] = df['Dur_min'].astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```
