DATA ENCODING

Machine learning models cannot  directly interpret categorical variables hence they require numerical representation
Therefore we need to convert categorical features into numerical values. This process is called data encoding 

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## create a simple dataframe
df = pd.DataFrame({'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
## create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
## perform fit and transform
## they have sorted in alphabetical of blue, then green and then red
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## converting the new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```
